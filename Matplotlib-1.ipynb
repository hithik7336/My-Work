```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
```

```python
plt.plot(x,y)

plt.xlabel('X-Axis')

plt.ylabel('Y-Axis')

plt.title('Title')
```

```python
plt.subplot(1,2,1)

plt.plot(x,y,'r')

plt.subplot(1,2,2)

plt.plot(y,x,'g')
```

```python
# OBJECT ORIENTED METHOD
```

```python
fig  = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X-Axis')

axes.set_ylabel('Y-Axis')

axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # L B W H

axes1.plot(x,y,'r')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes2.plot(y,x,'b')
```
