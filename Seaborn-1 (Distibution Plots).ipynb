```python
import numpy as np

import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')

pd.set_option('display.max_rows',1000)

pd.set_option('display.max_columns',1000)

pd.set_option('display.max_colwidth',1000)

pd.set_option('display.width',1000)
```

# DIST PLOT


```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

# JOINT PLOT

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

# PAIR PLOT 

```python
sns.pairplot(tips,hue='sex')
```

# RUG PLOT

```python
sns.rugplot(tips['total_bill'])
```

# KDE PLOT

```python
sns.kdeplot(tips['total_bill'])
```

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill'],color='green',bins=100)
```

```python
sns.distplot(tips['tip'],color='orange')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter',color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='purple')
```

```python
sns.pairplot(data=tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'],color='red')
```
