```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
x = sns.FacetGrid(data = tips, row = 'smoker', col = 'time')
x.map(sns.boxplot, 'total_bill' , 'day')
```

```python
x = sns.FacetGrid(data = tips, row = 'smoker', col = 'time')
x.map(plt.scatter, 'total_bill', 'tip')
```
